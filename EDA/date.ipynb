```python
import pandas as pd

ts_ms = pd.date_range(start='2025-01-01',
                    end='2025-10-02',
                    periods=100,
                    tz='Asia/Seoul')

data = { 'date': ts_ms.strftime('%Y-%m-%d')}
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df.date)
```

```python
# 년도를 추출해 새로운 칼럼으로 넣기
df['year'] = df.date.dt.year
df['month'] = df.date.dt.month
df['day_int'] = df.date.dt.weekday #0: 월요일

def convert_week(x):
    temp=['월','화','수','목','금','토','일']
    return temp[x]

df['day_str'] = df['day_int'].apply(lambda x: convert_week(x))
df.head()
```

```python
# 주말(토일)은 1, 나머지 0
def identify_weekend(x):
    if x=='토' or x=='일':
        return 1
    else:
        return 0

df['day_weekend'] = df['day_str'].apply(lambda x: identify_weekend(x))
df.head()
```

```python
import holidays

kr_holidays = holidays.KR(years=[2025])
# 기존 데이터에 추가하기
kr_holidays.append({'2025-10-10' : '휴강'})

for date, name in sorted(kr_holidays.items()):
    print(f"{date} : {name}")
```

```python
# holiday:1 , not:0
def holidays(x):
    if x in kr_holidays:
        return 1
    else:
        return 0

df['is_holiday'] = df['date'].apply(lambda x: holidays(x))
df.head()
```
