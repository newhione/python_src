```python
import pandas as pd
import seaborn as sns

flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.info() #category: 범위를 갖고있음
```

```python
flights.T
```

```python
flights.pivot(index='year', columns='month', values='passengers')
```

```python
# 연도별 승객수 변화 lineplot - sns
sns.lineplot(data=flights, x='year',y='passengers')
```

```python
# sns 시각화 - '신뢰구간'을 같이 표시하는 경우
# 신뢰구간: 데이터가 변동될 수 있는 구간 
sns.lineplot(data=flights, x='year',y='passengers', errorbar=('ci',95)) #95% 신뢰수준, 실선->평균 추세 / 음영->신뢰구간
# sns.lineplot(data=flights, x='year',y='passengers', errorbar=None) #신뢰구간 제거

# 신뢰구간이 넓으면 -> 불확실성이 크다 (데이터 분산이 크거나 샘플 수가 적은 경우)
# 신뢰구간이 좁으면 -> 안정적 (분산이 작거나 샘플 수가 많은 경우)
```

```python
# 월별로 그룹화
sns.lineplot(data=flights, x='year',y='passengers', hue='month')
```

```python
# 원하는 월만 뽑아서 보기
# isin()
sns.lineplot(
    data=flights[flights['month'].isin(['Jan', 'Mar', 'Dec'])],
    x='year', y='passengers', hue='month'
)
```
